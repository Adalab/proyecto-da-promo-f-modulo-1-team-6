# Preguntas y Respuestas


- Reglas del Juego de "Preguntas y Respuestas de Geografía":
    - El juego consiste en hacer preguntas de geografía y que el jugador adivine la respuesta correcta.

    - Cada pregunta tiene una respuesta única y correcta.

    - El jugador tiene un número limitado de intentos para responder correctamente a cada pregunta.

    - El juego sigue hasta que el jugador responda incorrectamente 3 de veces o haya respondido 5 preguntas correctamente.


- Pistas para las alumnas:

    1. Comienza creando un diccionario que contenga preguntas de geografía como claves y respuestas como valores.

    2. Implementa una función para seleccionar una pregunta aleatoria del diccionario.

    4. Dentro del bucle, presenta la pregunta al jugador y permita que ingresen sus respuestas.

    5. Lleva un seguimiento del número de preguntas realizadas y el número de respuestas correctas.

    6. Determina las condiciones de victoria y derrota del juego y sal del bucle cuando se cumplan estas condiciones.

```python
import random

# Diccionario de preguntas y respuestas
preguntas_respuestas = {
    "¿Cuál es la capital de Francia?": "París",
    "¿En qué año se fundó Apple Inc.?": "1976",
    "¿Cuál es el planeta más grande del sistema solar?": "Júpiter",
    "¿Quién escribió 'Romeo y Julieta'?": "William Shakespeare",
    "¿Cuál es el elemento químico con símbolo 'H'?": "Hidrógeno"
}

# Obtener una lista de preguntas
preguntas = list(preguntas_respuestas.keys())

# Inicializar la puntuación del jugador y el contador de fallos
puntuacion = 0
fallos = 0

# Inicializar una lista para llevar un registro de las preguntas realizadas
preguntas_realizadas = []

while puntuacion < 5 and fallos < 3:  # El juego se ejecuta hasta que el jugador alcance 5 puntos o cometa 3 fallos
    # Elegir una pregunta aleatoria que no se haya realizado
    pregunta = random.choice(preguntas)
    while pregunta in preguntas_realizadas:
        pregunta = random.choice(preguntas)
    
    preguntas_realizadas.append(pregunta)
    
    # Obtener la respuesta correcta
    respuesta_correcta = preguntas_respuestas[pregunta]
    
    # Mostrar la pregunta y obtener la respuesta del jugador
    print("Pregunta:", pregunta)
    respuesta_usuario = input(f"{pregunta} Tu respuesta: ").strip()
    
    # Comprobar si la respuesta es correcta
    if respuesta_usuario.lower() == respuesta_correcta.lower():
        print("¡Respuesta correcta!")
        puntuacion += 1  # Incrementar la puntuación en 1 punto
        print("Puntuación actual:", puntuacion)
    else:
        print(f"Respuesta incorrecta. La respuesta correcta es: {respuesta_correcta}")
        fallos += 1  # Incrementar el contador de fallos en 1
    
print("¡Juego terminado!")
if puntuacion == 5:
    print("¡Felicidades! Has alcanzado 5 puntos y has ganado el juego.")
else:
    print("Has cometido 3 fallos. Mejor suerte la próxima vez.")

```
