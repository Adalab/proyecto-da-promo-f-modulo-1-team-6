# Preguntas y Respuestas


- Reglas del Juego de "Preguntas y Respuestas de Geografía":
    - El juego consiste en hacer preguntas de geografía y que el jugador adivine la respuesta correcta.

    - Cada pregunta tiene una respuesta única y correcta.

    - El jugador tiene un número limitado de intentos para responder correctamente a cada pregunta.

    - El juego sigue hasta que el jugador responda incorrectamente 3 de veces o haya respondido 5 preguntas correctamente.


- Pistas para las alumnas:

    1. Comienza creando un diccionario que contenga preguntas de geografía como claves y respuestas como valores.

    2. Implementa una función para seleccionar una pregunta aleatoria del diccionario.

    3. Utiliza un bucle `while` para hacer preguntas de geografía hasta que el jugador responda un número específico de preguntas correctamente o alcance un número máximo de respuestas incorrectas.

    4. Dentro del bucle, presenta la pregunta al jugador y permita que ingresen sus respuestas.

    5. Lleva un seguimiento del número de preguntas realizadas y el número de respuestas correctas.

    6. Determina las condiciones de victoria y derrota del juego y sal del bucle cuando se cumplan estas condiciones.

```python
import random
#Función de instrucciones

def mostrar_instrucciones(jugador):
    print(f"Hola, {jugador}, vamos a jugar a Preguntas y respuestas de Geografía. Ganarás si consigues 5 puntos.")

#Función leer el diccionario

def obtener_respuesta_pregunta(diccionario, pregunta):
    return diccionario[pregunta]

#Función jugar al juego, 

def jugar_preguntas_geografia(diccionario, oportunidades):
    puntuacion = 0  #La puntuación empieza desde 0
    fallos = 3  #Solo se puede fallar 3 veces
    preguntas_realizadas = []  #Acumula las preguntas ya realizadas

#Se inica bucla while con las preguntas para el jugador

    while puntuacion < 5 and oportunidades > 0 and fallos > 0:
        pregunta = random.choice(list(diccionario.keys())) #Usamos random.choice ya que las preguntas las acumulamos en una lista. Obtenemos preguntas
        respuesta = obtener_respuesta_pregunta(diccionario, pregunta) #Obtenemos las respuestas
        jugador_respuesta = input(f"{pregunta} Tu respuesta: ").strip() #Preguntamos al usuario
        print(pregunta)  #Printemos pregunta
        preguntas_realizadas.append(pregunta) #Acumula las ya realizadas

        if jugador_respuesta.lower() == respuesta.lower():
            print("¡Correcto! Has ganado 1 punto")
            print("                                     ")
            puntuacion += 1
            print(f"Tienes {puntuacion} puntos")
            print("--------------------------------------")
        else:
            fallos -= 1
            oportunidades -= 1
            print(f"Incorrecto. La respuesta correcta es: {respuesta}. Solo puedes fallar {fallos} veces más.")
            print("                                      ")
            print(f"Tienes {oportunidades} oportunidades más para completar el juego")
            print("                                      ")
            print(f"Tienes {puntuacion} puntos")
            print("--------------------------------------")

        del diccionario[pregunta] #Eliminamos las preguntas si se repite, pero si vuelves a jugar si se repite

    return puntuacion  #Acabamos el bucle

jugador = input("Nombre del jugador: ")
mostrar_instrucciones(jugador)  #Llamamos a la funcion

juego = input("¿Preparado para jugar? S/N: ").upper() #Preguntamos al jugador

# Este bucle

while juego == "S":
    print("¡Vamos allá!\n") #Empieza el juego

#Diccionario

    diccionario = { 
        "¿Cuál es el río más largo del mundo?": "Nilo",
        "¿Cuál es la capital de Honduras?": "Tegucigalpa",
        "¿Cuál es el lago con mayor altitud?": "Titicaca",
        "¿Cuál es la isla más extensa?": "Groenlandia",
        "¿Cuántos mares existen en el mundo?": "60",
        "¿Cuántos océanos tiene la tierra?": "5",
        "¿Qué país tiene más habitantes?": "China",
        "¿Cuál es el desierto más grande del mundo?": "Sahara",
        "¿Qué país se encuentra entre dos continentes?": "Turquia",
        "¿Cual es la coordillera más larga del mundo?": "Los Andes",
        "¿Cuántos países forman la UE?": "27",
        "¿Cuál es la capital más septentrional del mundo?": "Reykjavik",
        "¿Cuál es la capital más austral del mundo?": "Wellington"
    }
#El jugador tiene 5 oportunidades para jugar

    oportunidades = 5
    puntuacion = jugar_preguntas_geografia(diccionario, oportunidades)


#Final del juego, pregunta si quieres volver a jugar y vuelves al 2º bucle, y este entra en la funcion de jugar_preguntas_geografia

    if puntuacion == 5:
        print(f"¡Enhorabuena, {jugador}! Has ganado.")
    else:
        print(f"Lo siento, {jugador}, has perdido :(.")
    
    juego = input("¿Quieres jugar otra vez? S/N: ").upper()


```
