```python
import random

jugador = input("Nombre del jugador: ")
print("Hola,", jugador, "Vamos a jugar a Preguntas y respuestas de Geografía. Ganarás si consigues 5 puntos.")

juego = input("¿Preparade para jugar? S/N").upper()

diccionario = {
    "¿Cuál es el río más largo del mundo?": "Nilo",
    "¿Cuál es la capital de Honduras?": "Tegucigualpa",
    "¿Cuál es el lago con mayor altitud?": "Titicaca",
    "¿Cuál es la isla más extensa?": "Groenlandia",
    "¿Cuántos mares existen en el mundo?": "60"
}

preguntas = list(diccionario.keys())

puntuacion = 0
oportunidades = 5
fallos = 3

while juego == "S" and (puntuacion < 5 and oportunidades > 0 and fallos > 0):
    pregunta_aleatoria = random.choice(preguntas)
    print(pregunta_aleatoria)

    jugador_respuesta = input("Tu respuesta: ").strip().lower()

    if jugador_respuesta == diccionario[pregunta_aleatoria].lower():
        print("¡Correcto! Has ganado 1 punto.")
        puntuacion += 1
        preguntas.remove(pregunta_aleatoria)
        print(f"Tienes {puntuacion} puntos.")
        print("--------------------------------------")
    else:
        fallos -= 1
        oportunidades -= 1
        print(f"Incorrecto! Solo puedes fallar {fallos} veces más.")
        print(f"Tienes {oportunidades} oportunidades más para completar el juego.")
        print(f"Tienes {puntuacion} puntos.")
        print("--------------------------------------")

    if puntuacion == 5:
        print(f"¡Enhorabuena, {jugador}! Has ganado.")
        juego = input("¿Quieres jugar otra vez? S/N").upper()
        oportunidades = 5
        fallos = 3
        puntuacion = 0

    elif oportunidades == 0 or fallos == 0:
        print(f"Lo siento, {jugador}, has perdido :(.") 
        juego = input("¿Quieres jugar otra vez? S/N").upper()
        oportunidades = 5
        fallos = 3
        puntuacion = 0
```

```python
#Define a dictionary of questions and answers.
qa_pairs = {
    "What is the capital of France?": "Paris",
    "Who was the first President of the United States?": "George Washington",
    "What is the chemical symbol for water?": "H2O",
    "Which 1975 film directed by Steven Spielberg features a man-eating great white shark?": "Jaws",
    "What is the largest continent by land area?": "Asia"
}
# Function to ask questions and check answers.
def play_game():
    score = 0
    for question, answer in qa_pairs.items():
        user_answer = input(question + " ").strip()# quita el espacio al principio y al final del input
        if user_answer.lower() == answer.lower():
            print("Correct! You earned 10 points.")
            score += 10
        else:
            print("Incorrect. The correct answer is: " + answer)
    print("Your final score: " + str(score))
# Main game loop.
while True:
    print("Welcome to the Question-Answer Game!")
    play_game()
```
