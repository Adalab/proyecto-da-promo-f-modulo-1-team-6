- Reglas del Juego "Tic-Tac-Toe":

    - El juego se juega en un tablero de 3x3.

    - Dos jugadores, uno representa "X" y el otro "O", alternan turnos.

    - El objetivo es obtener tres símbolos del mismo tipo en línea horizontal, vertical o diagonal.

    - El juego termina cuando un jugador gana o el tablero se llena (empate).

    - El jugador que logre colocar tres de sus símbolos en línea es el ganador.


- Pistas para las alumnas:

    2. Implementa una función para imprimir el tablero en la pantalla en un formato claro y legible.

    3. Utiliza un bucle `while` para alternar los turnos de los jugadores y recopilar sus movimientos.

    4. Verifica después de cada movimiento si hay un ganador o si el juego termina en empate.

    5. Implementa una función que verifique si un jugador ha ganado alineando tres símbolos en fila, columna o diagonal.

    6. Asegurate de validar que una casilla seleccionada por un jugador no esté ocupada antes de permitir el movimiento.
    

```python
jugador1 = 'X'
jugador2 = 'O'

juego = True


tablero = [[' ',' ',' '], [' ',' ',' '], [' ',' ',' ']]


# funcion para printear el tablero siempre
def imprimir_tablero(tablero):
    for fila in tablero:
        print("-------------")
        hueco = '| ' # '| ' es para que saque la barrita al principio y en general para que tenga en cuenta el espacio vacío donde vamos a meter las X y O
        espacio = ' | ' # para que separe los espacios
        for posicion in fila:
            hueco += posicion + espacio # en cada jugada se meten X y O en la posicion que elija el jugador (espacio no varía) y se van sumando hasta que termine el bucle con todos los huecos llenos
        print(hueco)
    print("-------------\n")

# función para comprobar que el tablero no está lleno, y si lo está, hay empate
def tablero_completo(tablero, juego): # dentro del tablero
    for fila in tablero: #tenemos tres filas
        for hueco in fila: #y¡ y dentro de cada fila tenemos 3 huecos
            if hueco == ' ': # los huevos están vacíos y se van llenando según avanza el juego
                return False #es falso que esté completo, así que hasta que eso pase, sigue funcionando el bucle
            
    print('El tablero está completo, hay un empate.')
    # cuando el tablero sí esté completo, que no queden huecos, entonces se cerará el bucle
    return True # así cerraríamos el bucle
    

# funcion para comprobar que las combinaciones horizontales, verticales y diagonales coinciden sin ser huecos en blanco teniendo en cuenta las variables de los dos jugadores
def comprobar_combinaciones(tablero, juego, jugador1, jugador2):

    if tablero[0][0] != ' ' and tablero[1][1] != ' ' and tablero[2][2] != ' ':
        print('¡Has ganado!\n')
    
        return juego == False
        
    elif tablero[0][2] != ' ' and tablero[1][1] != ' ' and tablero[2][0] != ' ':
        print('¡Has ganado!\n')
    
        return juego == False

    elif tablero[0][0] != ' ' and tablero[0][1] != ' ' and tablero[0][2] != ' ':
        print('¡Has ganado!\n')
    
        return juego == False

    elif tablero[1][0] != ' ' and tablero[1][1] != ' ' and tablero[1][2] != ' ':
        print('¡Has ganado!\n')
    
        return juego == False

    elif tablero[2][0] != ' ' and tablero[2][1] != ' ' and tablero[2][2] != ' ':
        print('¡Has ganado!\n')

        return juego == False

    elif tablero[0][0] != ' ' and tablero[1][0] != ' ' and tablero[2][0] != ' ':
        print('¡Has ganado!\n')
    
        return juego == False

    elif tablero[0][1] != ' ' and tablero[1][1] != ' ' and tablero[2][1] != ' ':
        print('¡Has ganado!\n')
    
        return juego == False

    elif tablero[0][2] != ' ' and tablero[1][2] != ' ' and tablero[2][2] != ' ':
        print('¡Has ganado!\n')
    
        return juego == False
    


# aquí es que empieza el juego realmente
# después de crear todas las funciones y variables que van a intervenir


print('¡Bienvenidos! Vamos a jugar al Tic, tac, toe o 3 en raya.')
print('¿Preparades?\n')

imprimir_tablero(tablero)

while juego:
    print('Jugador 1, es tu turno.')

    fila = int(input('Jugador 1, elige fila')) -1 
    columna = int(input('Jugador 1, elige columna'))-1

    # lo del -1 es porque como los índices empiezan en 0 y estamos contando con el imput 1, 2 y 3, así nos aseguramos de que lo que pone el usuario y lo que lee el programa el el tablero coincide

    if tablero[fila][columna] != ' ':
        print('Esta posición está ocupada, elige otra.') #Si se elige una posición ocupada, no te deja, pedimos que elijas otra vez
        imprimir_tablero(tablero) #y sacamos el tablero para que se vea bien, teniendo en cuenta que a veces se queda el print enganchado, esto podría ser también una solución a eso

        fila = int(input('Jugador 1, elige fila')) -1 #se vuelve a elegir posición
        columna = int(input('Jugador 1, elige columna'))-1

    tablero[fila][columna] = jugador1 #y se guarda la posición elegida por cada jugador, en este caso, el jugador1, en el tablero


    imprimir_tablero(tablero)

    juego = tablero_completo(tablero, juego) != True

    comprobar_combinaciones(tablero, juego, jugador1, jugador2)

    
    


    print('Jugador 2, te toca.')


    fila = int(input('Jugador 2, elige fila')) -1
    columna = int(input('Jugador 2, elige columna'))-1

    if tablero[fila][columna] != ' ':
        print('Esta posición está ocupada, elige otra.')
        fila = int(input('Jugador 2, elige fila')) -1
        columna = int(input('Jugador 2, elige columna'))-1

    tablero[fila][columna] = jugador2


    imprimir_tablero(tablero)

    juego = tablero_completo(tablero, juego) != True

    comprobar_combinaciones(tablero, juego, jugador1, jugador2)
```
